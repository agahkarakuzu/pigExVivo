```python
# Read masked voxels of Heart 2 for each method per week 
# hr2Data.mat is given by pigExVivo.m

mat = loadmat('hr2Data.mat', struct_as_record=False)

volSasha = np.squeeze(np.array(mat['sashaSer']))
volMolli = np.squeeze(np.array(mat['molliSer']))
volShmolli = np.squeeze(np.array(mat['shmolliSer']))
volIR = np.squeeze(np.array(mat['irSer']))


def reverse_colourmap(cmap, name = 'my_cmap_r'):
    
    reverse = []
    k = []   

    for key in cmap._segmentdata:    
        k.append(key)
        channel = cmap._segmentdata[key]
        data = []

        for t in channel:                    
            data.append((1-t[0],t[2],t[1]))            
        reverse.append(sorted(data))    

    LinearL = dict(zip(k,reverse))
    my_cmap_r = mpl.colors.LinearSegmentedColormap(name, LinearL) 
    return my_cmap_r

cmap = cm.gist_stern
cmap_r = reverse_colourmap(cmap)


def f(Week):
    x = Week
    clr = cmap_r
    plt.figure(figsize=(8, 8))
   
    if x<4:
        mn = 700
        mx = 1400
    else:
        mn = 400
        mx = 1200
      
 
    ax4 = plt.subplot(141)
    imgplot = plt.imshow(volIR[:,:,x-1])
    imgplot.set_cmap(clr)
    imgplot.set_clim(mn,mx)
    ax4.set_xticks([])
    ax4.set_yticks([])
    ax4.set_title('Ground Truth IR')
    ax1 = plt.subplot(142)
    imgplot = plt.imshow(volSasha[:,:,x-1])
    imgplot.set_cmap(clr)
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.set_title('SASHA')
    imgplot.set_clim(mn,mx)
    ax2 = plt.subplot(143)
    imgplot = plt.imshow(volMolli[:,:,x-1])
    imgplot.set_cmap(clr)
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.set_title('MOLLI')
    imgplot.set_clim(mn,mx)
    ax3 = plt.subplot(144)
    imgplot = plt.imshow(volShmolli[:,:,x-1])
    imgplot.set_cmap(clr)
    imgplot.set_clim(mn,mx)
    ax3.set_xticks([])
    ax3.set_yticks([])
    ax3.set_title('SHMOLLI')
    cax = plt.axes([0.99, 0.38, 0.025, 0.35])
    plt.colorbar(cax=cax)
    plt.tight_layout()
    plt.clim(mn, mx)
    plt.title('T1 (Heart 2)')
    plt.show()  
    return Week

interact(f, Week =widgets.IntSlider(min=1,max=6,step=1,value=1))

# Get box element for each masked voxel clusters 
def getBoxHr2(inpt,week,color,method):
    week = go.Box(
        y = inpt,
        name = method,
        jitter = 0.3,
        pointpos = -1.8,
        boxpoints = 'all',
        marker = dict(
            color = color,
            size = 2,
        ),
        line = dict(
            color = color)
        
    )
    return week

shmHr2 = mat['shmolliHr2']
molHr2 = mat['molliHr2']
sasHr2 = mat['sashaHr2']
irHr2  = mat['irHr2']

# Get all box elements for a given week 
def getWeekHr2(week,shmHr2,molHr2,sasHr2,irHr2):
    bx1 = getBoxHr2(irHr2[0,week-1].vec,week,irColor,'IR')
    bx2 = getBoxHr2(sasHr2[0,week-1].vec,week,sashaColor,'SASHA')
    bx3 = getBoxHr2(molHr2[0,week-1].vec,week,molliColor, 'MOLLI')
    bx4 = getBoxHr2(shmHr2[0,week-1].vec,week,shmolliColor, 'SHMOLLI')
    
    return [bx1, bx2, bx3, bx4]

# Loop through weeks to get all box elements 
for i in range(1,7):
    exec('wk' + str(i) + '= getWeekHr2(' + str(i) + ',shmHr2,molHr2,sasHr2,irHr2)')
    
# This is the collected data of Heart 2 for each method for all weeks 
data = wk1 + wk2 + wk3 + wk4 + wk5 + wk6

# This loop will give boolean labels to be used by updatemenus. This labels are vectors of 24 elements
# 4 Methods X 6 Weeks following an order [1(ir) 1(sasha) 1(molli) 1(shmolli) 0 0 0 0 ...] for week 1

wk1Label = []
wk2Label = []
wk3Label = []
wk4Label = []
wk5Label = []
wk6Label = []


for i in range(0,6):
    for j in range(0,4):
        
        if i == 0:
            wk1Label.append(True)
        else:
            wk1Label.append(False)
        if i == 1:
            wk2Label.append(True)
        else:
            wk2Label.append(False)
        if i == 2:
            wk3Label.append(True)
        else:
            wk3Label.append(False)
        if i == 3:
            wk4Label.append(True)
        else:
            wk4Label.append(False)
        if i == 4:
            wk5Label.append(True)
        else:
            wk5Label.append(False)
        if i == 5:
            wk6Label.append(True)
        else:
            wk6Label.append(False)
        

updatemenus = list([
    dict(
         
         buttons=list([   
            dict(label = 'Week 1',
                 method = 'update',
                 args = [{'visible': wk1Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 1 (Prefixation)</i>',
                          'yaxis.range': (376,1537),
                          'bgcolor': wk1color,
                          'annotations': []}]),
             dict(label = 'Week 2',
                 method = 'update',
                 args = [{'visible': wk2Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 2 (Prefixation)</i>',
                          'yaxis.range': (376,1537),
                          'annotations': []}]),
            dict(label = 'Week 3',
                 method = 'update',
                 args = [{'visible': wk3Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 3 (Prefixation)</i>',
                          'yaxis.range': (376,1537),
                          'annotations': []}]),
             dict(label = 'Week 4',
                 method = 'update',
                 args = [{'visible': wk4Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 4 (Postfixation)</i>',
                          'yaxis.range': (376,1537),
                          'annotations': []}]),
            dict(label = 'Week 5',
                 method = 'update',
                 args = [{'visible': wk5Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 5 (Postfixation)</i>',
                          'yaxis.range': (376,1537),
                          'annotations': []}]),
            dict(label = 'Week 6',
                 method = 'update',
                 args = [{'visible': wk6Label},
                         {'title': '<b>Myocardial T1 values for Heart2</b> <br> <i>Voxelwise distribution on Week 6 (Postfixation)</i>',
                          'yaxis.range': (376,1537),
                          'annotations': []}])
            
        
        ]),
        type = 'buttons',
        bgcolor='#EDEEEE',
        font=dict(
            
           
            color= '#0A2463'
        ),
    
        active= -1,
        showactive = True,
        
    )
])

layout = go.Layout(
    width=990,
    height=600,
    showlegend=False, title =  'Please select a week (from the buttons on the left) to start interactive display.')

layout['updatemenus'] = updatemenus



fig = dict(data=data, layout=layout)

#py.iplot(fig, filename='hr2Weekly')
```
